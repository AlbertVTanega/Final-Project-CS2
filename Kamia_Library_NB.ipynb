```python
borrowers_ref = db.reference("borrowers")
borrowers_data = borrowers_ref.get()

print("UNRETURNED BOOKS:\n")

for info in borrowers_data:

    if info is None:
        continue

    name = info["student"]
    books = info["books"]

    for book in books:
        if book.get("return_date") is None or book.get("return_date") == "":

            print("Student:", name)
            print("Book:", book["book_title"])
            print("Borrow Date:", book["borrow_date"])
            print("                                            ")
```
