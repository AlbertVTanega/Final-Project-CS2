```python
import firebase_admin
from firebase_admin import credentials, db

cred = credentials.Certificate("/content/firebase_key.json")

# Check if Firebase app is already initialized to avoid re-initialization errors
if not firebase_admin._apps:
    firebase_admin.initialize_app(cred, {
        "databaseURL": "https://kamia-tanega-library-db-43079-default-rtdb.asia-southeast1.firebasedatabase.app/"
    })

ref = db.reference("library")

while True:
    print("\n===== LIBRARY MANAGEMENT MENU =====")
    print("1. Display Borrowers and Their Books")
    print("2. Add Borrower")
    print("3. Update Borrower Info")
    print("4. Delete Borrower")
    print("5. Show Unreturned Books")
    print("6. Exit")

    choice = input("Enter choice: ")

    # 1. DISPLAY
    if choice == "1":
        borrowers = ref.get()
        if borrowers:
            print("\nBorrower List:")
            for key, b in borrowers.items():
                print(f"ID: {b['borrower_id']}, Student: {b['student']}")
                if 'books' in b:
                    for book in b['books']:
                        print(f"  - {book['book_title']} (Author: {book['author']})")
        else:
            print("No borrowers found.")

    # 2. ADD
    elif choice == "2":
        bid = input("Enter Borrower ID: ")
        name = input("Enter Student Name: ")

        # Adding a starter book to match your JSON list style
        book_t = input("Enter first book title: ")
        author = input("Enter author: ")

        new_borrower = {
            "borrower_id": int(bid),
            "student": name,
            "books": [
                {
                    "book_title": book_t,
                    "author": author,
                    "borrow_date": "2026-03-18",
                    "return_date": None
                }
            ]
        }

        ref.child(bid).set(new_borrower)
        print("Borrower added successfully!")

    # 3. UPDATE
    elif choice == "3":
        bid = input("Enter ID of borrower to update: ")
        borrower_ref = ref.child(bid)
        data = borrower_ref.get()

        if data:
            new_name = input("Enter new name: :) ")
            borrower_ref.update({"student": new_name})
            print("Name updated successfully!")
        else:
            print("Borrower not found.")

    # 4. DELETE
    elif choice == "4":
        bid = input("Enter ID to delete: ")
        if ref.child(bid).get():
            ref.child(bid).delete()
            print("Borrower deleted successfully!")
        else:
            print("Borrower not found.")

    # 5. UNRETURNED
    elif choice == "5":
        borrowers = ref.get()
        if borrowers:
            print("\nUnreturned Books:")
            for key, b in borrowers.items():
                if 'books' in b:
                    for book in b['books']:
                        # Use .get() to safely check for 'return_date' to prevent KeyError
                        if book.get('return_date') is None:
                            print(f"{b['student']} has: {book['book_title']}")
        else:
            print("No borrowers found.")

    # 6. EXIT
    elif choice == "6":
        print("Exiting...")
        break
```
